```python
import numpy as np
import pandas as pd
```

```python
movies_df = pd.read_csv("movies.csv")
ratings_df = pd.read_csv("ratings.csv")
```

```python
movies_df
```

```python
ratings_df
```

```python
df = pd.merge(movies_df,ratings_df,on='movieId')
```

```python
df.head(50)
```

```python
df.drop("timestamp",axis=1,inplace=True)
```

```python
df.head()
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head(20)
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head(20)
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

```python
sns.jointplot(x='rating',y='num of ratings',data=ratings,alpha=0.5)
```

```python
df
```

```python
moviemat = df.pivot_table(index='userId',columns='title',values='rating')
moviemat.head()
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
Shawshank_Redemption_user_ratings = moviemat['Shawshank Redemption, The (1994)']
Pulp_Fiction_user_ratings = moviemat['Pulp Fiction (1994)']
Shawshank_Redemption_user_ratings.head()
```

```python
similar_to_Shawshank_Redemption = moviemat.corrwith(Shawshank_Redemption_user_ratings)
similar_to_Pulp_Fiction = moviemat.corrwith(Pulp_Fiction_user_ratings)
```

```python
similar_to_Shawshank_Redemption.sort_values(ascending=False)
```

```python
corr_Shawshank= pd.DataFrame(similar_to_Shawshank_Redemption,columns=['Correlation'])
corr_Shawshank.dropna(inplace=True)
corr_Shawshank.head()
```

```python
corr_Shawshank.sort_values('Correlation',ascending=False).head()
```

```python
corr_Shawshank = corr_Shawshank.join(ratings['num of ratings'])
corr_Shawshank.head()
```

```python
corr_Shawshank[corr_Shawshank["num of ratings"]>100].sort_values('Correlation',ascending=False).head(10)
```

```python
corr_Pulp_Fiction = pd.DataFrame(similar_to_Pulp_Fiction,columns=['Correlation'])
corr_Pulp_Fiction.dropna(inplace=True)
corr_Pulp_Fiction = corr_Pulp_Fiction.join(ratings['num of ratings'])
corr_Pulp_Fiction[corr_Pulp_Fiction['num of ratings']>100].sort_values('Correlation',ascending=False).head(10)
```
